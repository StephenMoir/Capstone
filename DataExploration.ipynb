## BREAST CANCER DATA EXPLORATION
    
#### Load the required libraries

```python
import io
import os
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
from scipy.stats import norm
```

```python
%matplotlib inline
```

#### Read in the data

```python
# read in the csv file
local_data = 'data/data.csv'

# print out some data
bc_df = pd.read_csv(local_data)
print('Data shape (rows, cols): ', bc_df.shape)
print()
bc_df.head()
```

```python
# features
features_list = bc_df.columns.values
print('Features: \n', features_list)
```

#### Check for missing data

```python
# drop the column of nulls
clean1_bc_df = bc_df.drop('Unnamed: 32', axis = 1)

print('Data shape (rows, cols): ', clean1_bc_df.shape)
clean1_bc_df.head()
```

```python
# drop any incomplete data
clean_bc_df = clean1_bc_df.dropna(axis=0)
print('Data shape (rows, cols): ', clean_bc_df.shape)
```

#### Create a numerical outcome variable

```python
clean_bc_df['diag_value'] = [1 if x == 'M' else 0 for x in bc_df['diagnosis']]
```

```python
clean_bc_df = clean_bc_df.drop('diagnosis', axis = 1)
```

```python
clean_bc_df.head()
```

#### Explore the variables

<b> a. Radius (mean of distances from center to points on the perimeter) </b><br>
a1. Mean <br>
a2. Standard Error <br>
a3. Largest (mean of 3 largest)

```python
# radius
radius_list = ['radius_mean', 'radius_se', 'radius_worst']
n_bins = 50 

for column_name in radius_list:
    data = clean_bc_df[column_name]
    ax=plt.subplots(figsize=(6,3))
    # get data by column_name and display a histogram
    ax = plt.hist(data, bins=n_bins, density=True, alpha=0.6, color='y')
    
    # normal distribution
    mu , std = norm.fit(data)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    plt.plot(x, p, 'k', linewidth=2)
    
    title="Fit results: mu = %.2f,  std = %.2f" % (mu, std) +" for " + column_name
    plt.title(title, fontsize=12)
    plt.show()
```

### Scale the data

As a general rule of thumb use <br>
    1. Standardisation when the data is approximately normally distributed <br>
    2. Min-Max scaling when it is not
<br>
<br>
Using the histograms above, it can be seen that, whilst not perfect in every case, a Standardisation can be justified
    

```python
# only scale the variables - not the id or outcome
var_bc_df = clean_bc_df.loc[:,'radius_mean':'fractal_dimension_worst']
var_bc_df.head()
```

```python
# scale the data to have mean = 0 and std = 1
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()



# store them in this dataframe
scaled_bc_df = pd.DataFrame(scaler.fit_transform(var_bc_df.astype(float)))

# get same features and State-County indices
scaled_bc_df.columns=var_bc_df.columns


scaled_bc_df.head()
```

Construct the scaled dataframe

```python
scaled_final_bc_df = pd.concat([clean_bc_df['id'], scaled_bc_df, clean_bc_df['diag_value']], axis = 1, sort = False)
scaled_final_bc_df.head()
```

Write to a csv file for use in PCA notebook

```python
scaled_data = 'data/scaled.csv'
scaled_final_bc_df.to_csv(scaled_data)
```
